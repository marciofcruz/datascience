```python
def km_media(dataset, ano_atual):
    auxiliar = {}
    for item in dataset.items():
        auxiliar.update({item[0] : item[1]['km'] / (ano_atual - item[1]['ano'])})
        #print(item)
        
    return auxiliar
        
print(km_media(dados, 2019))
```
